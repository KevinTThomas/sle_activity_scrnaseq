```python
banchereau_module_file = project_dir / "banchereau_modules.json"
```

```python
import json
```

```python
with banchereau_module_file.open("r") as bmf:
    banchereau_modules = json.load(fp=bmf)
```

```python
adata.obs["M1.2"] = sc.tl.pca(adata[:,[_ for _ in banchereau_modules["M1.2"] if _ in adata.var.index]], n_comps=1, copy=True).obsm["X_pca"]
adata.obs["M3.4"] = sc.tl.pca(adata[:,[_ for _ in banchereau_modules["M3.4"] if _ in adata.var.index]], n_comps=1, copy=True).obsm["X_pca"]
adata.obs["5.12"] = sc.tl.pca(adata[:,[_ for _ in banchereau_modules["M5.12"] if _ in adata.var.index]], n_comps=1, copy=True).obsm["X_pca"]
```

```python
found_isgs = (
    [_ for _ in banchereau_modules["M1.2"] if _ in adata.var.index] +
    [_ for _ in banchereau_modules["M3.4"] if _ in adata.var.index] +
    [_ for _ in banchereau_modules["M5.12"] if _ in adata.var.index]
)
```

```python
adata_exprs_df = pd.DataFrame(adata.X, index=adata.obs.index, columns=adata.var.index)
```

```python
adata.obs["M1.2_paper_score"] = adata_exprs_df.loc[:,[_ for _ in banchereau_modules["M1.2"] if _ in adata.var.index]].mean(1)
adata.obs["M3.4_paper_score"] = adata_exprs_df.loc[:,[_ for _ in banchereau_modules["M3.4"] if _ in adata.var.index]].mean(1)
adata.obs["M5.12_paper_score"] = adata_exprs_df.loc[:,[_ for _ in banchereau_modules["M5.12"] if _ in adata.var.index]].mean(1)
```

```python
adata.obs["paper_IFN_score"] = adata_exprs_df.loc[:,[_ for _ in banchereau_modules["M1.2"]+banchereau_modules["M3.4"]+banchereau_modules["M5.12"] if _ in adata.var.index]].mean(1)
```

```python
fig, ax = plt.subplots(figsize=(18,6))
sns.violinplot(x=adata.obs["paper_cell_type"], y=adata.obs["paper_IFN_score"], ax=ax)
sns.stripplot(x=adata.obs["paper_cell_type"], y=adata.obs["paper_IFN_score"], ax=ax, color="black", alpha=0.25, size=.75)
ax.tick_params(axis='x', rotation=90)
```

```python
adata.obs["type_disease"] = adata.obs["paper_cell_type"].astype("str") + adata.obs["disease_status"].astype("str")
adata.obs["type_disease"] = adata.obs["type_disease"].astype("category")
```
